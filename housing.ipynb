```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
#from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras import models
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
dataframe = pd.read_csv("E:\\Taha\\tensorflow\\allfiles\\housing.csv", delim_whitespace = True, header = None)
```

```python
dataset = dataframe.values
x = dataset[:,0:13]
y = dataset[:,13]
```

```python
from tensorflow.keras.utils import normalize
```

```python
x = normalize(x, axis = 1)
y = normalize(x, axis = 1)
```

```python
def baseline_model():
    model = models.Sequential()
    model.add(Flatten())
    model.add(Dense(13,))
    model.add(Dense(1,))
    model.compile(optimizer = 'Adam', loss = 'mse', metric = ['mse'])
    return model
```

```python
seed = 7
np.random.seed(seed)
```

```python
estimator = KerasRegressor(build_fn = baseline_model, batch_size = 5, epochs = 100, verbose = 1)
```

```python
kfold = KFold(n_splits = 10, random_state = seed)
results = cross_val_score(estimator, x, y, cv = kfold)
print("Results: %.2f(%.2f)MSE" % (results.mean(), results.std()))
```
