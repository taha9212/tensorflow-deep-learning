```python
import tensorflow as tf
```

```python
print(tf.__version__)
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(tranies_img, tranies_label),(test_img, test_label) = mnist.load_data()
```

```python
tranies_img = tf.keras.utils.normalize(tranies_img, axis = 1)
test_img = tf.keras.utils.normalize(test_img, axis = 1)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(512, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation = tf.nn.softmax))
```

```python
model.compile(optimizer = tf.keras.optimizers.RMSprop(),
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
model.fit(tranies_img, tranies_label, epochs = 3, batch_size = 128)
```

```python
model.evaluate(test_img, test_label)
```
